```python
import pandas as pd

penguins = sns.load_dataset("penguins")
```

```python
penguins.shape
```

```python
# =================================================-
#### Slide 9: Data preprocessing  ####

# Check for null values
penguins.isna().sum()
```

```python
# Drop NA's and reassign clean data
penguins = penguins.dropna()
penguins.isna().sum()
```

```python
penguins.head()
```

```python
sns.displot(
    data=penguins,
    x="body_mass_g",
    hue="species",
    palette=["darkorange", "teal", "blue"],
    kde=True,
    rug=True,
);
```

```python
# =================================================-
#### Slide 13: Plot a layered histogram  ####

# plt.figure(figsize=(6,4))     #<- set the figure size

sns.histplot(
    data=penguins,  # <- set data
    kde=True,
    x="body_mass_g",  # <- set x variable
    hue="species",  # <- grouping parameter
    palette="dark:m_r",
);  # <- set color
```

```python
penguins
```

```python
def grouped_bar(df, group):
    df.select_dtypes("number").assign(grouping_variable=penguins[group]).groupby("grouping_variable").mean().T.plot.barh();

grouped_bar(penguins, "species")
```

```python
# =================================================-
#### Slide 14: Univariate plots: box plot  ####

plt.figure(figsize=(5, 3))  # <- set the figure size

# Create a simple boxplot
sns.violinplot(
    data=penguins,  # <- set data
    x="body_mass_g",  # <- set y variable
    palette=["m"],
);  # <- set color
```

```python
# =================================================-
#### Slide 15: Plot a box plot  ####

plt.figure(figsize=(6, 4))  # <- set the figure size
sns.violinplot(
    data=penguins,  # <- set data
    y="species",  # <- set x variable
    x="body_mass_g",  # <- set y variable
    palette=["m", "cyan", "y"],
);  # <- set color
```

```python
# =================================================-
#### Slide 15: Plot a box plot  ####

plt.figure(figsize=(6, 4))  # <- set the figure size
sns.boxplot(
    data=penguins,  # <- set data
    y="species",  # <- set x variable
    x="body_mass_g",  # <- set y variable
    palette=["m", "cyan", "y"],
);  # <- set color
```

```python
# =================================================-
#### Slide 15: Plot a box plot  ####

plt.figure(figsize=(6, 4))  # <- set the figure size
sns.swarmplot(
    data=penguins,  # <- set data
    y="species",  # <- set x variable
    x="body_mass_g",  # <- set y variable
    hue="species",
);  # <- set color
```

```python
# =================================================-
#### Slide 16: Univariate plots: bar plot  ####

plt.figure(figsize=(6, 4))  # <- set the figure size

# Create a barplot
sns.violinplot(
    data=penguins,  # <- set the data
    y="sex",  # <- set x variable
    x="body_mass_g",  # <- set y variable
    # errorbar = "ci",
    palette="viridis",
);  # <- set color
```

```python
g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", s=100)
g.plot_marginals(sns.rugplot, height=.15)
g.plot_marginals(sns.histplot, kde=True);
```

```python
# =================================================-
#### Slide 22: Bivariate plots: residual plot  ####

sns.residplot(
    data=penguins,
    x="bill_length_mm",  # <- set x variable
    y="flipper_length_mm",  # <- set y variable
    lowess=True,
    color="b",
);
```

```python
import numpy as np
```

```python
# =================================================-
#### Slide 23: Multivariate plots: Heatmaps  ####
corr = penguins.select_dtypes("number").corr()
sns.heatmap(
    corr.iloc[1:, :-1],
    vmin=-1,
    vmax=1,
    center=0,
    annot=True,
    square=True,
    mask=np.triu(corr)[1:, :-1],
    annot_kws={"fontsize": 14},
    # cbar=False,
).tick_params(labelsize=12)
plt.xticks(rotation=45, ha="right", rotation_mode="anchor");
plt.yticks(rotation=45, ha="right", rotation_mode="anchor");
```

```python
# =================================================-
#### Slide 25: Format plots in seaborn (cont'd)  ####

g = sns.lmplot(
    data=penguins,
    x="bill_length_mm",  # <- set x variable
    y="flipper_length_mm",  # <- set x variable
    height=4,  # <- set the height
    fit_reg=True,
)
g.set(
    title="Bill Length versus Flipper Length",  # <- set the title
    xlabel="Bill Length (mm)",  # <- set label for x variable
    ylabel="Flipper Length (mm)",
);
```

```python
ax = penguins.groupby("species")["body_mass_g"].mean().sort_values().plot.barh()
for i in ax.containers:
    ax.bar_label(
        i,
    )
ax.set_xlim((0, 6000))
ax.set_ylabel("Species", rotation=0)
ax.set_xlabel("Body Mass (g)")
ax.yaxis.set_label_coords(-0.25, 0.47)
```
