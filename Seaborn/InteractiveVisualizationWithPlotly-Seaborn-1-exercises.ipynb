#### Task 4
##### Create a boxplot of `total_bill`.
##### Create a second boxplot of `total_bill` grouped by `day`.
#### Result:


```python
#=================================================-
#### Slide 15: Plot a box plot  ####

sns.boxplot(data = tips,    #<- set data
        y = "day",          #<- set x variable
        x = "total_bill",      #<- set y variable
        # palette = ["m", "cyan", "y"]
              );        #<- set color
```

```python
#=================================================-
#### Slide 15: Plot a box plot  ####

sns.swarmplot(
    data=tips,
    y="day",
    x="total_bill",
    linewidth=1,
    alpha=.4,
    edgecolor="k"
)
sns.violinplot(
    data=tips,
    y="day",
    x="total_bill",
);
```

#### Task 9
##### Create a heatmap of `total_bill` by day and time using the below code to make a pivot table.


```python
tips2 = tips.groupby(['day', 'time']).mean()[['total_bill']]
tips2.reset_index(inplace = True)
tips2 = tips2.pivot('day', 'time', 'total_bill')
```

```python
tips2
```

```python
sns.heatmap(
    tips2,
    annot=True,
    square=True,
    annot_kws={"fontsize":14},
).tick_params(labelsize=12)
plt.xticks(rotation=0);
plt.ylabel("Day", rotation=0)
plt.yticks(rotation=0);
```

#### Task 10
##### Use the code from Task 4 for the second boxplot and change the xlabel to `Days of the Week` and the ylabel to `Total Bill ($)` and the title to `Distributions of restaurant bills over the week`
#### Result:


```python
sns.swarmplot(
    data=tips,
    y="day",
    x="total_bill",
    linewidth=1,
    alpha=.4,
    edgecolor="k"
)
sns.violinplot(
    data=tips,
    y="day",
    x="total_bill",
)
plt.title("Distributions of restaurant bills over the week");
plt.xlabel("Total Bill ($)");
plt.ylabel("Days of the Week");
```

```python
sns.stripplot(
    data=tips,
    y="day",
    x="tip",
    hue="time",
    linewidth=1,
    alpha=.4,
    edgecolor="k"
)
sns.violinplot(
    data=tips,
    y="day",
    x="tip",
    hue="time",
).set(
    title="Distributions of restaurant bills over the week",
    xlabel="Total Bill ($)",
    ylabel="Days of the Week",
)
plt.legend(labels=["Lunch", "Dinner"]);
```
